```python
import feynman_path
import numpy as np
from qiskit import *
from qiskit.visualization import plot_histogram
```

## Before Adding CNOT Gate

```python
circ= QuantumCircuit(3)
circ.h(0)
circ.cx(0,1)
circ.x(0)
circ.z(1)
circ.h(1)
circ.barrier(0,1)
circ.z(0)
circ.h(0)
circ.cx(1,0)
circ.h(1)
circ.draw("mpl")
```

```python
n_qubits = 3
font = 12
ws_label = 4+0.55*n_qubits  # Label width relative to font size
w_time = 60+ws_label*font  # Diagram column width
```

```python
f = feynman_path.Diagram(
    n_qubits, font=font, ws_label=ws_label, w_time=w_time)
f.perform_h(0)
f.perform_cnot(0, 1)
f.perform_x(0)
f.perform_z(1)
f.perform_h(1)
f.perform_z(0)
f.perform_h(0)
f.perform_cnot(1, 0)
f.perform_h(1)
f.draw()
```

## After Adding CNOT Gate

```python
circ= QuantumCircuit(3)
circ.h(0)
circ.cx(0,1)
circ.x(0)
circ.z(1)
circ.h(1)
circ.barrier(0,1)
circ.cx(1,2) # Extera CNOT Gate
circ.z(0)
circ.h(0)
circ.cx(1,0)
circ.h(1)
circ.draw("mpl")
```

```python
f = feynman_path.Diagram(
    n_qubits, font=font, ws_label=ws_label, w_time=w_time)
f.perform_h(0)
f.perform_cnot(0, 1)
f.perform_x(0)
f.perform_z(1)
f.perform_cnot(1, 2, pre_latex=r'\color{red!80!black}') # Extera CNOT Gate
f.perform_h(1)
f.perform_z(0)
f.perform_h(0)
f.perform_cnot(1, 0)
f.perform_h(1)
f.draw()
```
