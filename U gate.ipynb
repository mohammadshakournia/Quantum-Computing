```python
import numpy as np
from qiskit import *
from qiskit.visualization import plot_bloch_multivector
from math import pi
```

## Part 2

#### Before Applying Gate

```python
circ1 = QuantumCircuit(2)
circ1.h(0)
statevector_backend1_before=Aer.get_backend('statevector_simulator')
final_state1_before=execute(circ1,statevector_backend1_before).result().get_statevector()
plot_bloch_multivector(final_state1_before,title="Before Applying U Gate")
```

#### Applying Gate

```python
circ1.cp(pi/4,0,1)
circ1.draw("mpl")
```

#### After Applying Gate

```python
statevector_backend1_after=Aer.get_backend('statevector_simulator')
final_state1_after=execute(circ1,statevector_backend1_after).result().get_statevector()
plot_bloch_multivector(final_state1_after,title="After Applying U Gate")
```

```python
if final_state1_after == final_state1_before:
    print("No Changes in Qubits have been occurred")
else:
    print("Before:")
    print(final_state1_before)
    print("After:")
    print(final_state1_after)
```
