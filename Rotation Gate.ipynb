```python
import numpy as np
from qiskit import *
from qiskit.circuit import Parameter
from qiskit.visualization import plot_bloch_multivector
%matplotlib inline
```

```python
qc = QuantumCircuit(1)
plot_bloch_multivector(qc,title="Before Applying Rx Gate")
```

```python
qc.rx(np.pi,0)
qc.draw("mpl")
```

```python
plot_bloch_multivector(qc,title="After Applying Rx Gate")
```

```python
backend = Aer.get_backend('unitary_simulator')
job = execute(qc, backend)
job.result().get_unitary(qc, decimals=1)
```
